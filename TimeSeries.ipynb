```python
import pandas as pd
```

```python
df = pd.read_csv("https://www.datahub.io/core/co2-ppm/r/co2-mm-mlo.csv")
```

```python
df['Date']
```

```python
pd.to_datetime(df['Date'])
```

```python
df['datetime'] = pd.to_datetime(df['Date']) 
```

```python
df

# We can see the new datetime column we have created
```

```python
import seaborn as sns
```

```python
sns.lineplot(data=df, x="datetime", y="Interpolated")
sns.lineplot(data=df, x="datetime", y="Trend")
```
